```python
from pathlib import Path

import pandas as pd
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

from trainer import train_model

N_SPLITS = 3
OUTDIR = Path("out/")
FEATURE_FILE = Path("data/feature_5se_sasa_rd_df.json")
```

```python
# with flag mismatch, no af2_max_flag
train_model(
    model_cls=RandomForestClassifier,
    model_init_params={
        "n_estimators": 128,
        "max_depth": None,
        "min_samples_split": 2,
        "min_samples_leaf": 1,
        "max_features": None,
        "class_weight": "balanced",
        "bootstrap": True,
        "random_state": 42,
        "n_jobs": -1,
        "verbose": 0,
    },
    feature_df=pd.read_json(FEATURE_FILE),
    categ_cols=[
        "residue",
        "nbr_s1",
        "nbr_s2",
        "nbr_s3",
        "nbr_s4",
        "nbr_s5",
        "nbr_sasa_s1",
        "nbr_sasa_s2",
        "nbr_sasa_s3",
        "nbr_sasa_s4",
        "nbr_sasa_s5",
        "nbr_depth_s1",
        "nbr_depth_s2",
        "nbr_depth_s3",
        "nbr_depth_s4",
        "nbr_depth_s5",
        # "af2_max_flag",
    ],
    numer_cols=["af2_sd", "af2_max_ss%", "af2_flag_repeat_len", "max_flag_mismatch", "plddt"],
    target_col="nmr_max_flag_clf",
    af2_pred_col="af2_max_flag_clf",
    n_splits=N_SPLITS,
    split_by_protein=True,
    outdir=OUTDIR / "rf-5s-exc-sasa-rd-clf/",
)
# Check result on secondary structure mismatch vs no mismatch
# Mismatch should have lower plddt
```
