```python
from matplotlib.colors import LogNorm


def plot_graphs(filename: Path):
    res = pd.read_csv(filename)

    y_true = res.y_true
    y_pred = res.y_pred.clip(upper=1)
    y_af2 = res.y_pred_af2

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    n_bins = 175
    cmap = ["magma", "plasma", "inferno", "cubehelix"][2]
    norm = LogNorm()

    h1 = ax1.hexbin(y_true, y_af2, gridsize=n_bins, cmap=cmap, norm=norm)
    fig.colorbar(h1, ax=ax1)
    ax1.set_xlabel("True LDDT")
    ax1.set_ylabel("AF2 pLDDT")
    ax1.set_title("True vs AF2 pLDDT")
    ax1.set_xlim(left=0)
    ax1.set_ylim(bottom=0)
    ax1.grid(True, alpha=0.2)

    h2 = ax2.hexbin(y_true, y_pred, gridsize=n_bins, cmap=cmap, norm=norm)
    fig.colorbar(h2, ax=ax2)
    ax2.set_xlabel("True LDDT")
    ax2.set_ylabel("Model Predicted LDDT")
    ax2.set_title("True vs Model Prediction")
    ax2.set_xlim(left=0)
    ax2.set_ylim(bottom=0)
    ax2.grid(True, alpha=0.2)

    fig.tight_layout()
    plt.show()


file = Path("/home/fahim/compbio/out/rf-5s-exclusive/fold_0/predict.csv")
plot_graphs(file)
```
